```python
import pandas as pd

def add_moving_average(df):
    # Ensure the 'Date' column is in datetime format
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Sort the DataFrame by 'Date' to ensure the moving average is calculated in the correct order
    df = df.sort_values(by='Date')
    
    # Calculate the moving average using a window size of 7
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
    
    return df

# Example usage:
# Create a sample DataFrame
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05', '2023-01-06', '2023-01-07', '2023-01-08', '2023-01-09'],
    'Sales': [100, 200, 150, 300, 250, 400, 350, 450, 500]
}
df = pd.DataFrame(data)

# Add the MovingAverage column
df = add_moving_average(df)

print(df)
```
